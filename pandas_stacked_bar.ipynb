```python
df=pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/pisa2015.csv',index_col='Country')
```

```python
df
```

```python
df[:5].plot(kind='barh',stacked=True,title='Pisa score 2015');
```

```python
df[:5].plot(kind='barh',stacked=True,title='Pisa score 2015',figsize=(8,4));
```

```python
ax=df[:5][['Maths','Science']].plot(kind='barh',stacked=True,color=['gold','c'],
                                    title='Pisa Score 2015');
ax.set_xlabel('Score')
ax.set_ylabel('')
```

```python
df['Total'] = df.Maths+df.Science+df.Reading
df
```

```python
df.nlargest(10,'Total')[['Maths','Science','Reading']].plot(kind='barh',stacked=True)
```

```python
df.loc[['Japan','Taiwan','Vietnam','Thailand'],['Maths','Science','Reading']].plot(kind='barh',stacked=True)
```

```python
ax=df.loc[['Japan','Taiwan','Vietnam','Thailand'],['Maths','Science','Reading']].plot(kind='barh',stacked=True)
ax.xaxis.set_major_formatter(FuncFormatter(lambda v,p:"{:,.0f}".format(v))) #set comma delemiter
ax.set_xlabel('Score')
ax.set_ylabel('');
```
