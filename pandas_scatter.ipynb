```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df.sample(5)
```

```python
df.plot.scatter(x='displ',y='cty')
```

```python
%config InvalidIndexError.figue_format = 'retina'
df.plot.scatter(x='displ',y='cty');
```

```python
df.plot.scatter(x='displ',y='cty',grid ='True') #Add Grid
```

```python
df.plot.scatter(x='displ',y='cty',grid ='True',title='City miles per gallon vs. engine size (litres)')
```

```python
df.plot.scatter(x='displ',y='cty',
                grid ='True',
                title='City miles per gallon vs. engine size (litres)')
```

```python
df.plot.scatter(x='displ',y='cty',
                grid ='True',
                c=df.drv.map({'4':'green','f':'blue','r':'red'}), #Add color by drv
                title='City miles per gallon vs. engine size (litres)')
```

```python
df.plot.scatter(x='displ',y='cty',
                grid ='True',
                c=df.drv.map({'4':'green','f':'blue','r':'red'}), #Add color by drv
                s=df.displ*10,
                title='City miles per gallon vs. engine size (litres)')
```

```python
df.plot.scatter(x='displ',y='cty',
                grid ='True',
                c=df.drv.map({'4':'green','f':'blue','r':'red'}), #Add color by drv
                s=df.cyl*10,
                title='City miles per gallon vs. engine size (litres)')
```

```python
df.plot.scatter(x='displ',y='cty',
                grid ='True',
                c=df.drv.map({'4':'green','f':'blue','r':'red'}), #Add color by drv
                s=df.cyl*10,
                alpha=.5,
                title='City miles per gallon vs. engine size (litres)')
```

```python
df.plot.scatter(x='displ',y='cty',
                grid ='True',
                c=df.drv.map({'4':'green','f':'blue','r':'red'}), #Add color by drv
                s=30,
                alpha=.5,
                title='City miles per gallon vs. engine size (litres)')
```

```python
df.plot.scatter(x='displ',y='cty',
                grid ='True',
                c=df.drv.map({'4':'green','f':'blue','r':'red'}), #Add color by drv
                s=30,
                alpha=.5,
                ylim=(0,40), #Set Start YLIM
                title='City miles per gallon vs. engine size (litres)')
```

```python
df.plot.scatter(x='displ',y='cty',
                grid ='True',
                c=df.drv.map({'4':'green','f':'blue','r':'red'}), #Add color by drv
                s=30,
                alpha=.5,
                ylim=(0,40), #Set Start YLIM
                xlim=(0,10), #Set Start XLIM
                title='City miles per gallon vs. engine size (litres)')
```
