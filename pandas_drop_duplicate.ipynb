```python
import pandas as pd
pd.__version__
```

```python
url='https://github.com/prasertcbs/tutorial/raw/master/duplicates.csv'
df=pd.read_csv(url)
df
```

```python
df.drop_duplicates() #drop ข้อมูลซ้ำทุกคอลัมน์
```

```python
df.drop_duplicates(subset='manufacturer')
```

```python
df.drop_duplicates(subset='model')
```

```python
df.drop_duplicates(subset=['model','manufacturer'])
```

```python
df.drop_duplicates(subset=['year','class'])
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.drop_duplicates(subset=['manufacturer','model'],inplace=True)
df
```
